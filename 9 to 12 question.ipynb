```python
def char_count(string:str):
    if not isinstance(string,str):
        raise TypeError("Input must be string")
    if len(string) == 0:
        raise ValueError("input can not be empty")
    char_count = {}
    for char in string.lower().replace(" ",""):
        if char in char_count:
            char_count[char]+=1
        else:
            char_count[char] =1
    return char_count 

if __name__ == "__main__":
    text = "python"
    try:
        result = char_count(text)
        print(result)
    except (TypeError,ValueError) as e:
        print(e)
    except Exception as e:
        print(e)
        
```

```python
def factorial (n):
    if n==0 or n==1:
        return 1
    else:
        return n*factorial(n-1)
```

```python
def remove_duplicat(list1 : list) -> int:
    if not isinstance(list1,list):
        raise TypeError("input must be a list")
    for i in list1:
        if not isinstance(i,(int,float)):
            raise ValueError("all elemente must be integer")
    list2 = []
    for i in list1:
        if i not in list2:
            list2.append(i)
    return list2 
if __name__ == "__main__":
    try:
        user_input = input("enter a number seprate by comma")
        number = [int(i.strip()) for i in user_input.split(",")]
        result = remove_duplicat(number)
        print(f"List without duplicate {result}")
       
    except (TypeError,ValueError) as e:
        print(e)
    except Exception as e:
        print(e)
       
```
